```python
from transformers import AutoTokenizer, AutoModel
import torch
```

```python
model_name = "bert-base-uncased"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModel.from_pretrained(model_name)
```

```python
def get_bert_embeddings(tokens):
    # Tokenize the text
    tokenized_text = tokenizer.tokenize(' '.join(tokens))

    # Truncate or split text if it exceeds BERT model's max token limit
    max_token_limit = tokenizer.max_model_input_sizes[model_name]
    if len(tokenized_text) > max_token_limit:
        tokenized_text = tokenized_text[:max_token_limit]

    if not tokenized_text:
        # Handle the case where tokenized_text is empty
        return None

    # Convert tokens to input IDs
    input_ids = tokenizer.convert_tokens_to_ids(tokenized_text)

    # Create a tensor from the input IDs
    inputs = torch.tensor(input_ids).unsqueeze(0)  # Add batch dimension

    # Get the BERT embeddings for the text
    with torch.no_grad():
        outputs = model(inputs)

    # Extract the embeddings from the BERT model
    embeddings = outputs[0].mean(dim=1)  # Use mean pooling

    return embeddings
```

```python
def array_to_string(arr):
    if isinstance(arr, np.ndarray):
        return np.array2string(arr, separator=',', precision=2)  # You can adjust separator and precision as needed
    else:
        return str(arr)
```

```python
import re
```

```python
def extract_numeric(cell):
    try:
        # Use regular expressions to extract numeric values
        numeric_values = re.findall(r'[-+]?\d*\.\d+|\d+', cell)
        # Convert the first extracted value to a float
        if numeric_values:
             return float(numeric_values[0])
        else:
            return None  # Handle the case where no numeric values are found
    except (ValueError, IndexError):
        return None
```
